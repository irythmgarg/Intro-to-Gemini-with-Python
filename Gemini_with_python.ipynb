```python
# Importing Markdown display function from IPython to render formatted text in Jupyter Notebooks.
from IPython.display import Markdown

# Importing textwrap to help with text formatting and indentation.
import textwrap

# Function to convert plain text into nicely formatted Markdown for better readability in notebooks.
def to_markdown(text):
    # Replace bullet points ('•') with asterisks ('*') to follow Markdown syntax for unordered lists.
    text = text.replace('•', ' *')

    # Indent every line of the text with '> ' to render it as a blockquote in Markdown.
    # 'predicate=lambda line: True' ensures that every line is indented.
    return Markdown(textwrap.indent(text, '> ', predicate=lambda line: True))
```

```python
## demo
sample_text = "• Point one • Point two"
display(to_markdown(sample_text))
```

```python
# Importing the Google Generative AI SDK to interact with generative models.
import google.generativeai as genai

# Retrieving the Google API key securely from the user data (stored in Colab environment).
GOOGLE_API_KEY = userdata.get('api_key')

# Configuring the Generative AI SDK with the provided API key to authenticate requests.
genai.configure(api_key=GOOGLE_API_KEY)

# Initializing the model with the specified model ID ("models/gemma-3-27b-it" is a placeholder).
model = genai.GenerativeModel("models/gemma-3-27b-it")

# Making a request to the model to generate content based on the prompt "Who is Sachin?".
response = model.generate_content("Who is Sachin?")
```

```python
# Accessing the text content of the response generated by the model.
# This will return the output or response text from the model's generate_content method.
response.text
```

```python
# Accessing the candidates attribute from the response object.
# This contains a list of possible generated outputs (candidates) from the model for the given prompt.
response.candidates
```

```python
# Importing the necessary libraries for displaying images and interacting with the file system.
import os
from IPython.display import Image, display

# URL of the image to be downloaded
image_url = "https://stimg.cardekho.com/images/carexteriorimages/630x420/BMW/M5-2025/11821/1719462197562/front-left-side-47.jpg?impolicy=resize&imwidth=480"

# Using curl (command-line tool) to download the image from the given URL and save it locally as "car_image.jpg".
os.system(f"curl -o car_image.jpg {image_url}")

# Displaying the downloaded image in the notebook using IPython's Image function.
display(Image(filename="car_image.jpg"))
```

```python
# Importing the Image class from the Python Imaging Library (PIL) to open and manipulate images.
from PIL import Image

# Using curl to download the image from the specified URL and save it as "car_image.jpg".
os.system(f"curl -o car_image.jpg \"{image_url}\"")

# Opening the downloaded image using PIL to process it for the generative model.
img = Image.open("car_image.jpg")

# Step 4: Create a Gemini multimodal model to process both text and image inputs.
# The model "gemini-1.5-flash" is chosen for multimodal tasks (text + image generation).
model2 = genai.GenerativeModel("gemini-1.5-flash")  # As of now, use gemini-1.5-* for multimodal

# Step 5: Generate content by providing both a text prompt and the image.
# The model will generate a description of the car based on the given prompt and the image.
response = model2.generate_content(["Describe this car in detail:", img])
```

```python
# Accessing the text content of the response generated by the model.
# This will return the output or response text from the model's generate_content method.
response.text
```

```python
# Calling the `to_markdown` function to convert the generated content (response.text) into a Markdown-formatted block.
# This will render the generated content in a more readable and formatted way in a Jupyter notebook.
to_markdown(response.text)
```
